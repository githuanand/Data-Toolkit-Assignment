8. Create a line plot with multiple lines using Matplotlib



```python
# Importing matplotlib
import matplotlib.pyplot as plt

# Data for plotting
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [25, 20, 15, 10, 5]

# Creating a line plot with multiple lines
plt.plot(x, y1, label='y = x^2', color='blue')
plt.plot(x, y2, label='y = 30 - x^2', color='red')

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot Example')

# Displaying the legend
plt.legend()

# Displaying the plot
plt.show()
```

9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold

```python
# Importing pandas
import pandas as pd

# Creating a sample DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 40]}

df = pd.DataFrame(data)

# Filtering rows where 'Age' is greater than 30
filtered_df = df[df['Age'] > 30]

# Displaying the result
print(filtered_df)
```

10. Create a histogram using Seaborn to visualize a distribution

```python
# Importing seaborn and matplotlib
import seaborn as sns
import matplotlib.pyplot as plt

# Creating a sample DataFrame
data = {'Age': [22, 25, 28, 29, 32, 35, 38, 40, 42, 45]}

df = pd.DataFrame(data)

# Creating a histogram
sns.histplot(df['Age'], kde=True)

# Adding labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')

# Displaying the plot
plt.show()
```

11. Perform matrix multiplication using NumPy


```python
# Importing numpy
import numpy as np

# Creating two matrices
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])

# Performing matrix multiplication
result_matrix = np.dot(matrix1, matrix2)

# Displaying the result
print("Result of matrix multiplication:\n", result_matrix)
```

13. Create a 3D scatter plot using Plotly

```python
# Importing Plotly
import plotly.express as px

# Creating a sample DataFrame for 3D plot
data = {'X': [1, 2, 3, 4, 5],
        'Y': [1, 4, 9, 16, 25],
        'Z': [2, 3, 4, 5, 6]}

df = pd.DataFrame(data)

# Creating a 3D scatter plot
fig = px.scatter_3d(df, x='X', y='Y', z='Z', title='3D Scatter Plot Example')

# Displaying the plot
fig.show()
```
